```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from datetime import datetime
```

```python
past_data = pd.read_excel('Past data.xlsx')
```

```python
future_data = pd.read_excel("May future.xlsx")
```

```python
past_data.head()
```

```python
future_data.head()
```

```python
def splitdate(x):
    spli = x.split(" ")
    return spli[0]
```

```python
future_data['Date'] = future_data['Date'].apply(splitdate)
past_data['Date'] = past_data['Date'].apply(splitdate)
```

```python
# data['Date'] = pd.to_datetime(data['Date'],format='%d%b%Y')
past_data.set_index('Date',inplace=True)
future_data.set_index('Date',inplace=True)
```

```python
# data.drop('Room Revenue',axis=1,inplace=True)
past_data.drop('Guests',axis=1,inplace=True)
future_data.drop('Guests',axis=1,inplace=True)
```

```python
past_data2 = past_data.drop(['Trans Rm Rev','Grp Rm Rev','Grp ADR','Trans ADR'],axis=1)
future_data2 = future_data.drop(['Trans Rm Rev','Grp Rm Rev','Grp ADR','Trans ADR'],axis=1)
```
